```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

When visualizing individual variables, it is important to undestand the data types we are dealing with

```python
print(df.dtypes)
```

Calculating correlation between variables of type 'int64' or 'float64' using the method 'corr':

```python
df.corr()
```

Diagonal elements are always one because it is correlation of a data type with itself.

```python
df[['bore', 'stroke', 'compression-ratio', 'horsepower']].corr()
```

It is always good to visualize continuous numerical variables using scatterplots with fitted lines. 'regplot' plots the scatterplot along with the fitted regression line for the data.

```python
# Engine size as potential predictor variable of price
sns.regplot(x = 'engine-size', y = 'price', data = df)
plt.ylim(0,)
```
