Calculating correlation between variables of type 'int64' or 'float64' using the method 'corr':

```python
df.corr()
```

Diagonal elements are always one because it is correlation of a data type with itself.

```python
df[['bore', 'stroke', 'compression-ratio', 'horsepower']].corr()
```

It is always good to visualize continuous numerical variables using scatterplots with fitted lines. 'regplot' plots the scatterplot along with the fitted regression line for the data.

```python
# Engine size as potential predictor variable of price
sns.regplot(x = 'engine-size', y = 'price', data = df)
plt.ylim(0,)
```

There is a positive direct correlation betweem these two variables. Engine size is a good predictor of price since the regression line is almost a perfect diagonal line

```python
df[['engine-size', 'price']].corr()
```

```python
# Highway mpg is a potential predictor variable for price.
sns.regplot(x = 'highway-mpg', y = 'price', data = df)
```

As the highway-mpg goes up, the price goes down, which suggests an inverse/negative relationship between the two variables

```python
df[['highway-mpg', 'price']].corr()
```

```python
sns.regplot(x  = 'peak-rpm', y = 'price', data = df)
```

All data points are scatterd and far from the fitted regression line and thus is not a reliable variable for prediction

```python
df[['peak-rpm', 'price']].corr()
```

```python
df[['stroke', 'price']].corr()
```

```python
sns.regplot(x = 'stroke', y = 'price', data = df)
```

### Descriptive Statistical Analysis

'describe' function computes the basic statistics for all continuous variables. Nan values are skipped

- count of variable
- mean
- standard deviation
- minimum value
- interquartile range or IQR
- maximum value

By default, all variables of type object are skipped

```python
df.describe()
```

```python
df.describe(include = ['object'])
```

```python
df['drive-wheels'].value_counts()
```

```python
df['drive-wheels'].value_counts().to_frame()
```

```python
driveWheelsCounts = df['drive-wheels'].value_counts().to_frame()
driveWheelsCounts.rename(columns = {'drive-wheels' : 'value_counts'}, inplace = True)
driveWheelsCounts
```

```python
driveWheelsCounts.index.name = 'drive-wheels'
driveWheelsCounts
```

```python
# Repeating the process with engine-location as a variable
engineLocationCount = df['engine-location'].value_counts().to_frame()
engineLocationCount.rename(columns = {'engine-location' : 'value-counts'}, inplace = True)
engineLocationCount.index.name = 'engine-location'
engineLocationCount.head(10)
```

Only 3 cars with rear engine-location and 198 cars with front engine-location. Class imbalance thus not a good predictor variable for price

### Basic of Grouping

'groupby' method groups data by different categories. Analysis is performed on individual groups

```python
df['drive-wheels'].unique()
```
